```python
from queue import PriorityQueue

class Nodo:
  def __init__ (self, caracter, prob):
    self.caracter = caracter
    self.prob = prob
    self.tag = -1
    self.rigth = None
    self.left = None

class Arbol:
  def __init__ (self, raiz ):
    self.raiz = raiz

  # Add comparison method for PriorityQueue
  def __lt__(self, other):
    return self.raiz.prob < other.raiz.prob

  def get_codigos (self, raiz, codigo="", codigos=None):
    if codigos is None:
      codigos = {}
    if raiz:
      if raiz.caracter is not None:
        codigos[raiz.caracter] = codigo
      self.get_codigos(raiz.left, codigo + "0", codigos)
      self.get_codigos(raiz.rigth, codigo + "1", codigos)
    return codigos

def min_probs ( arboles ):
  min_1 = arboles.get()[1]
  min_2 = arboles.get()[1]
  return [min_1, min_2]

# Construir arbol de Huffman
def construir_arbol(frecuencias):
  pq = PriorityQueue()
  for char, prob in frecuencias.items():
    nodo = Nodo(char, prob)
    pq.put((prob, Arbol(nodo)))

  while pq.qsize() > 1:
    prob1, arbol1 = pq.get()
    prob2, arbol2 = pq.get()
    nuevo = Nodo(None, prob1 + prob2)
    nuevo.left = arbol1.raiz
    nuevo.rigth = arbol2.raiz
    pq.put((nuevo.prob, Arbol(nuevo)))

  return pq.get()[1]


# Codificar texto
def codificar(texto, codigos):
  encoded_text = []
  for c in texto:
    if c in codigos:
      encoded_text.append(codigos[c])
    else:
      # Handle characters not in the codes dictionary (e.g., skip them)
      pass # Or you could add a default code, or raise a different error
  return ''.join(encoded_text)


# Guardar en binario
def guardar_binario(cadena, archivo):
  b = int(cadena, 2)
  n_bytes = (b.bit_length() + 7) // 8
  with open(archivo, 'wb') as f:
    f.write(b.to_bytes(n_bytes, 'big'))

# Leer binario
def leer_binario(archivo):
  with open(archivo, 'rb') as f:
    data = f.read()
  b = int.from_bytes(data, 'big')
  return bin(b)[2:]

# Decodificar usando el arbol
def decodificar(binario, raiz):
  resultado = []
  nodo = raiz
  for bit in binario:
    if bit == '0':
      nodo = nodo.left
    else:
      nodo = nodo.rigth
    if nodo.caracter is not None:
      resultado.append(nodo.caracter)
      nodo = raiz
  return ''.join(resultado)

# Mostrar arbol de forma legible
def mostrar_arbol(nodo, nivel=0, prefijo=""):
    if nodo is not None:
        print("    " * nivel + prefijo + (nodo.caracter if nodo.caracter is not None else f"({nodo.prob:.2f})"))
        mostrar_arbol(nodo.left, nivel + 1, "├── 0: ")
        mostrar_arbol(nodo.rigth, nivel + 1, "└── 1: ")

def menu():
  archivo_txt = "text.txt"
  with open(archivo_txt, 'r', encoding='utf-8') as f:
    texto = f.read()

  frecuencias = {}
  for c in texto:
    frecuencias[c] = frecuencias.get(c, 0) + 1

  total = sum(frecuencias.values())
  arbol = construir_arbol(frecuencias)
  codigos = arbol.get_codigos(arbol.raiz)

  while True:
    print("\n--- MENU ---")
    print("1. Codificar y decodificar archivo")
    print("2. Mostrar lista de caracteres, ocurrencias, probabilidades y codigos")
    print("3. Codificar texto ingresado por el usuario")
    print("4. Decodificar binario ingresado por el usuario")
    print("5. Mostrar arbol")
    print("6. Probar palabras (escalera, sitio, pie, suelo)")
    print("7. Salir")

    op = input("Opcion: ")

    if op == '1':
      codificado = codificar(texto, codigos)
      guardar_binario(codificado, 'salida.bin')
      print("Archivo codificado guardado en salida.bin")
      binario_leido = leer_binario('salida.bin')
      decodificado = decodificar(binario_leido, arbol.raiz)
      print("Texto decodificado (primeros 200 chars):")
      print(decodificado[:200])

    elif op == '2':
      for c, fr in frecuencias.items():
        prob = fr / total
        print(f"'{c}': {fr} ocurrencias, p={prob:.4f}, codigo={codigos[c]}")

    elif op == '3':
      user_text = input("Texto a codificar: ")
      print("Codificado:", codificar(user_text, codigos))

    elif op == '4':
      binario_user = input("Binario a decodificar: ")
      print("Decodificado:", decodificar(binario_user, arbol.raiz))

    elif op == '5':
      mostrar_arbol(arbol.raiz)

    elif op == '6':
      for palabra in ["escalera", "sitio", "pie", "suelo"]:
        print(palabra, ":", codificar(palabra, codigos))

    elif op == '7':
      break

    else:
      print("Opcion no valida")

if __name__ == "__main__":
  menu()
```
