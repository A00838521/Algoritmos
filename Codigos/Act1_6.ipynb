```python
# Bruno Vázquez Espinoza A00838521
# Algoritmos Greedy

def greedy_coin_collecting(filename): # Complejidad O(n)
    with open(filename, "r") as f:
        n = int(f.readline().strip())
        grid = []
        for _ in range(n):
            row = list(map(int, f.readline().strip().split()))
            grid.append(row)

    path = [(0, 0)]
    coins_collected = grid[0][0]
    i, j = 0, 0

    while i < n - 1 or j < n - 1:
        if i == n - 1:  # solo se puede mover a la derecha
            j += 1
        elif j == n - 1:  # solo se puede mover abajo
            i += 1
        else:
            # escoger la celda con más monedas
            if grid[i][j + 1] >= grid[i + 1][j]:
                j += 1
            else:
                i += 1

        coins_collected += grid[i][j]
        path.append((i, j))

    print("Greedy Coin Collecting")
    print("Monedas recolectadas:", coins_collected)
    print("Camino:", path)


def greedy_knapsack(filename): # Complejidad O(n log n)
    with open(filename, "r") as f:
        first_line = f.readline().strip().split()
        n, W = int(first_line[0]), int(first_line[1])
        items = []
        for _ in range(n):
            value, weight = map(int, f.readline().strip().split())
            items.append((value, weight, value / weight))

    # ordenar por ratio valor/peso
    items.sort(key=lambda x: x[2], reverse=True)

    total_value = 0
    total_weight = 0
    chosen_items = []

    for value, weight, ratio in items:
        if total_weight + weight <= W:
            chosen_items.append((value, weight))
            total_value += value
            total_weight += weight

    print("\nGreedy Knapsack")
    print("Capacidad máxima:", W)
    print("Valor total:", total_value)
    print("Peso total:", total_weight)
    print("Objetos elegidos (valor, peso):", chosen_items)


if __name__ == "__main__":
    greedy_coin_collecting("coins-n100.txt")
    greedy_knapsack("f3_l-d_kp_4_20.txt")
```

#¿Por qué es greedy?

- En CoinCollecting, el robot siempre escoge en cada paso el camino que parece mejor en ese momento (entre derecha o abajo, elige la casilla con más monedas). Eso es greedy porque no piensa en el futuro, solo toma la decisión local que se ve más prometedora.

- En Knapsack, también es greedy porque tomamos primero los objetos que tienen mejor relación valor/peso y vamos llenando la mochila mientras se pueda. Igual que el otro, no revisa combinaciones, solo se queda con lo que parece más conveniente al instante.
