```python
# 1. Determina cuales son los dos puntos mas cercanos dadas sus coordenadas x, y
import math

coor = []
with open("puntos-n100.txt", "r") as f:
  # número de puntos
  n = int(f.readline().strip())
  for _ in range(n):
    x, y = map(float, f.readline().split())
    coor.append([x, y])

# Inicializo variables
min_dist = float("inf")
p1 = 0
p2 = 0

# Se recorre todos los pares posibles
for i in range(len(coor)):
  for j in range(i + 1, len(coor)):
    dist = math.sqrt((coor[i][0] - coor[j][0])**2 + (coor[i][1] - coor[j][1])**2)

    if dist < min_dist:
      min_dist = dist
      p1, p2 = coor[i], coor[j]

print("Los puntos más cercanos son:", p1, "y", p2)
print("Distancia mínima:", min_dist)

# Complejidad O(n^2)
```

```python
# 1. Determina los indices donde aparece cada patron en una cadena
#    usando comparaciones entre caracteres unitarios.
#    La salida debe mostrar inicio y final de cada ocurrencia.

# Leer texto
with open("string-matching-Texto.txt", "r", encoding="utf-8") as f:
  cadena = f.read()

# Leer patrones
with open("string-matching-Patrones.txt", "r", encoding="utf-8") as f:
  patrones = [line.strip() for line in f if line.strip()]

print("Texto analizado:\n")
print(cadena)
print("\n--- Resultados ---\n")

for patron in patrones:
  print(f"Buscando: '{patron}'")
  encontrado = False
  for i in range(len(cadena) - len(patron) + 1):
    match = True
    for j in range(len(patron)):
      if cadena[i + j] != patron[j]:
        match = False
        break
    if match:
      print("Indices:", i, i + len(patron))
      encontrado = True
  if not encontrado:
    print("No encontrado")
  print()

# Complejidad O(n^2)
```

```python
# Algoritmo Permutaciones
# Computa las permutaciones de n, en orden lexicografico
# Complejidad O(n!)

def permutaciones_lex(n):
  p = list(range(1, n+1))
  L = [p.copy()]

  while True:
    i = -1
    for k in range(len(p)-2, -1, -1):
      if p[k] < p[k+1]:
        i = k
        break
    if i == -1:
      break

    j = -1
    for k in range(len(p)-1, i, -1):
      if p[i] < p[k]:
        j = k
        break

    p[i], p[j] = p[j], p[i]
    p[i+1:] = reversed(p[i+1:])
    L.append(p.copy())

  return L

print("Permutaciones en orden lexicográfico (n=3):")
for perm in permutaciones_lex(3):
    print(perm)
```

```python
# Algoritmo powerSetBin
# Computa recursivamente el conjunto potencia de n elementos
# Complejidad O(n*2^n)

def power_set_bin(n):
  if n == 1:
    return [[0], [1]]
  else:
    L1 = power_set_bin(n-1)
    L2 = list(reversed(L1))
    L1 = [[0] + subset for subset in L1]
    L2 = [[1] + subset for subset in L2]
    return L1 + L2

print("\nSubsets en binario (n=3):")
for subset in power_set_bin(3):
  print(subset)
```

```python
# Algoritmo de Maximo Subarray
# Dado un arreglo de enteros (positivos y negativos), encuentra el subarreglo contiguo con la suma máxima

# Pseudocodigo
# Input: Arreglo A de tamaño n
# Output: suma maxima de un subarreglo contiguo
# maxSuma = -infinito
# for i = 0 to n-1 do
#     suma = 0
#     for j = i to n-1 do
#         suma = suma + A[j]
#         if suma > maxSuma then
#             maxSuma = suma
# return maxSuma

# Complejidad O(n^2)

def max_subarray_bruteforce(A):
  max_suma = float("-inf")
  n = len(A)
  for i in range(n):
    suma = 0
    for j in range(i, n):
      suma += A[j]
      if suma > max_suma:
        max_suma = suma
  return max_suma

A = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
print("Arreglo:", A)
print("Máxima suma de subarreglo:", max_subarray_bruteforce(A))

# Hay manera de hacerlo mas eficiente con Divide y venceras O(n log n) dividiendo el arreglo en dos mitades, resolviendo recursivamente y combinando.
```
