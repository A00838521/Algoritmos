```python
c = [1,2,5,10]

# Complejidad O(2^n)
def coin_change(coins, n, k):
  if k == 0:
    return 1
  elif n <= 0 or k < 0:
    return 0
  else:
    return coin_change(coins, n, k - coins[n - 1]) + coin_change(coins, n - 1, k)

# Complejidad O(n^2)
def coins_tabu(c,n,k):
  T = [[0 for _ in range(k + 1)] for _ in range(n + 1)]
  T[0][0] = 1
  for i in range(1, n + 1):
    for j in range(k + 1):
      T[i][j] += T[i - 1][j]
      if (j - c[i - 1]) >= 0:
        T[i][j] += T[i][j - c[i - 1]]
  return T[n][k]

# Complejidad O(2^n)
def coins_memo(c,n,k,t):
  if k == 0:
    t[n][k] = 1
    return t[n][k]
  elif n <= 0 or k < 0:
    return 0
  elif t[n][k] != -1:
    return t[n][k]
  else:
    t[n][k] = coins_memo(c, n, k - c[n - 1], t) + coins_memo(c, n - 1, k, t)
    return t[n][k]
k = 10
memo_t = [[-1 for _ in range(k + 1)] for _ in range(len(c) + 1)]

print(coin_change(c, len(c), k))
print(coins_tabu(c, len(c), k))
print(coins_memo(c, len(c), k, memo_t))
```
