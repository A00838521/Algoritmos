```python
# Algoritmo de N-Queens con Backtracking Iterativo
# Complejidad peor caso: O(n!) porque en el peor de los casos se exploran todas las posibles colocaciones de reinas.

def imprimir_tablero(solucion, n):
  for fila in range(n):
    linea = ""
    for col in range(n):
      if solucion[fila] == col:
        linea += "R  "
      else:
        linea += "-  "
    print(linea.strip())
  print()


def n_queens_iterativo(n):
  soluciones = []
  fila = 0
  columnas = [-1] * n  # columnas[fila] = col de la reina en esa fila
  while fila >= 0:
    columnas[fila] += 1  # intentamos siguiente columna
    while columnas[fila] < n and not es_valida(columnas, fila):
      columnas[fila] += 1

    if columnas[fila] < n:
      if fila == n - 1:  # solución encontrada
        soluciones.append(columnas[:])
      else:
        fila += 1
        columnas[fila] = -1
    else:  # no hay columna válida
      columnas[fila] = -1
      fila -= 1

  return soluciones


def es_valida(columnas, fila_actual):
  for i in range(fila_actual):
    # Checa misma columna o diagonales
    if columnas[i] == columnas[fila_actual] or abs(columnas[i] - columnas[fila_actual]) == fila_actual - i:
      return False
  return True


if __name__ == "__main__":
  n = 5
  soluciones = n_queens_iterativo(n)
  print(f"\nNúmero total de soluciones: {len(soluciones)}\n")
  for s in soluciones:
    imprimir_tablero(s, n)

# ¿como se podría mejorar para llegar a una solución completamente definida mas rápido?

# Se podría mejorar el algoritmo usando otra estructura de datos como una cola de prioridad o un árbol.
# Esto serviría para organizar mejor las soluciones hijas y explorar primero las más prometedoras,
# lo que reduciría el tiempo en encontrar una solución completa en lugar de revisar a ciegas todas las combinaciones.
```
