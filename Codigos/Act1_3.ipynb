```python
def recolectar_monedas(matriz, n):
    # dp[i][j] = máximo de monedas que se pueden recolectar hasta la celda (i,j)
    dp = [[0] * n for _ in range(n)]

    dp[0][0] = matriz[0][0]

    for col in range(1, n):
        dp[0][col] = dp[0][col - 1] + matriz[0][col]

    for fila in range(1, n):
        dp[fila][0] = dp[fila - 1][0] + matriz[fila][0]
        for col in range(1, n):
            dp[fila][col] = max(dp[fila - 1][col], dp[fila][col - 1]) + matriz[fila][col]

    return dp[n - 1][n - 1]


if __name__ == "__main__":
    # Leer archivo de entrada
    with open("gold-n100.txt", "r") as archivo:
        lineas = archivo.readlines()

    n = int(lineas[0].strip())
    matriz_monedas = [list(map(int, linea.strip().split())) for linea in lineas[1:n+1]]

    resultado = recolectar_monedas(matriz_monedas, n)
    print("El número máximo de monedas que el robot puede recolectar es:", resultado)

# Complejidad: O(n^2)
# Relación de recurrencia: dp[i][j] = max(dp[i-1][j], dp[i][j-1]) + matriz[i][j]
# Esto funciona porque el robot solo puede venir desde arriba o izquierda, por lo que el valor máximo
# hasta (i,j) depende únicamente de esas dos celdas previas.
```
