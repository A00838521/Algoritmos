```python
import math
import matplotlib.pyplot as plt


# funcion de distancia
def dE(p1, p2):
  return math.sqrt( (p2[0] - p1[0])**2 + (p2[1] - p2[1])**2 )

```

```python
# Calculo del a, b, y c de la RECTA que corresponde al segmento
# Recibe un segmento, formado por dos puntos:  s es [ [x1, y1 ], [x2, y2] ]
def recta_abc( s ):
  # extrae las coordenadas de los puntos
  x1 = s[0][0]
  y1 = s[0][1]

  x2 =s[1][0]
  y2 =s[1][1]

  # Calcula a, b y c
  a = y2 - y1
  b = x1 - x2
  c = a*x1 + b*y1

  return (a, b, c)
```

```python
# A menos que sean paralelas, todo par de rectas intersecta en algun punto
def recta_interseccion( seg1, seg2 ):
	#  Obten el a, b y c de la recta que corresponde a cada segmento
  a1, b1, c1 = recta_abc( seg1 )
  a2, b2, c2 = recta_abc( seg2 )
  #  Calcula el determinante
  det = a1 * b2 - a2 * b1

  #  Usa el valor del determinate para decidir si se interceptan, o no, y donde
  try:
    x = (b2 * c1 - b1 * c2) / det
    y = (a1 * c2 - a2 * c1) / det
    return [x, y]
  except ZeroDivisionError:
    return None  # Las rectas son paralelas
```

```python
def segmento_interseccion(seg1, seg2):
  # Que las rectas se intersecten no significa que sus segmentos tambien
  p = recta_interseccion(seg1, seg2)
  if p is None:
    return None  # Las rectas son paralelas
  # Como saber si hay interseccion en los segmentos??
  for s in [seg1, seg2]:
    x_min = min(s[0][0], s[1][0])
    x_max = max(s[0][0], s[1][0])
    y_min = min(s[0][1], s[1][1])
    y_max = max(s[0][1], s[1][1])
    if not (x_min <= p[0] <= x_max and y_min <= p[1] <= y_max):
      return False  # No hay interseccion en los segmentos
  return p

```

```python
# Ejemplo para plot

# Recibe un segmento, formado por dos puntos:   [ [x1, y1 ], [x2, y2] ]
def plotSegment( segmento ):
  # Scatter sirve para dibujar puntos, funciona con listas
  # Sintaxis: plt.scatter( vals_x, vals_y )
  plt.scatter( [s[0] for s in segmento], [s[1] for s in segmento] )

  # Plot sirve para dibujar lineas, funciona con listas tambien
  # Sintaxis: plt.plot( vals_x, vals_y, opciones )
  plt.plot(  [s[0] for s in segmento], [s[1] for s in segmento] ,  linestyle="-")

  # Text sirve para agregar textos
  # Sintaxis:  plt.text( x, y, text, opciones)
  #plt.text(  ,  ,   , color="b" )


def writeInPlot( x, y, texto ):
  plt.text(x, y, texto)

# Caasos de prueba:

#	segmento 	p1       p2
S1 = [[1.0, 2.0],  [3.0, 4.0]]
S2 = [[2.0, 2.0],  [4.0, 1.0]]

interseccion = segmento_interseccion(S1,S2)
teoria = recta_interseccion(S1,S2)
if interseccion is None:
  writeInPlot((S1[0][0]+S1[1][0])/2, (S1[0][1]+S1[1][1])/2, "paralelas :(")
elif interseccion is False:
  writeInPlot((S1[0][0]+S1[1][0])/2, (S1[0][1]+S1[1][1])/2, "no se intersectan :(")
  writeInPlot(teoria[0], teoria[1], "posible interseccion :O")
  plt.scatter(teoria[0], teoria[1], color="y")
else:
  writeInPlot(interseccion[0], interseccion[1], "interseccion :O")
  plt.scatter(interseccion[0], interseccion[1], color="r")

plotSegment( S1)
plotSegment( S2)
writeInPlot(1.5, 1.5, "dos segmentos :O")
plt.show()

# Otros casos de prueba: con cruce
S1 = [[1.0, 2.0],  [3.0, 4.0]]
S2 = [[1.0, 3.0],  [4.0, 1.0]]

interseccion = segmento_interseccion(S1,S2)
teoria = recta_interseccion(S1,S2)
if interseccion is None:
  writeInPlot((S1[0][0]+S1[1][0])/2, (S1[0][1]+S1[1][1])/2, "paralelas :(")
elif interseccion is False:
  writeInPlot((S1[0][0]+S1[1][0])/2, (S1[0][1]+S1[1][1])/2, "no se intersectan :(")
  writeInPlot(teoria[0], teoria[1], "posible interseccion :O")
  plt.scatter(teoria[0], teoria[1], color="y")
else:
  writeInPlot(interseccion[0], interseccion[1], "interseccion :O")
  plt.scatter(interseccion[0], interseccion[1], color="r")


plotSegment( S1)
plotSegment( S2)
writeInPlot(1.5, 1.5, "dos segmentos :O")
plt.show()



# Otros casos de prueba: paralelas
S1 = [[1.0, 2.0],  [3.0, 4.0]]
S2 = [[1.0, 4.0],  [3.0, 6.0]]

interseccion = segmento_interseccion(S1,S2)
teoria = recta_interseccion(S1,S2)
if interseccion is None:
  writeInPlot((S1[0][0]+S1[1][0])/2, (S1[0][1]+S1[1][1])/2, "paralelas :(")
elif interseccion is False:
  writeInPlot((S1[0][0]+S1[1][0])/2, (S1[0][1]+S1[1][1])/2, "no se intersectan :(")
  writeInPlot(teoria[0], teoria[1], "posible interseccion :O")
  plt.scatter(teoria[0], teoria[1], color="y")
else:
  writeInPlot(interseccion[0], interseccion[1], "interseccion :O")
  plt.scatter(interseccion[0], interseccion[1], color="r")

plotSegment( S1)
plotSegment( S2)
writeInPlot(2.5, 2.5, "dos segmentos :O")
plt.show()

```
